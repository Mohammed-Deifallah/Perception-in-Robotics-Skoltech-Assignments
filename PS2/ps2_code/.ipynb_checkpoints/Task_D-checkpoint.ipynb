```python
import numpy as np
import matplotlib.pyplot as plt

input_file = 'evaluation-input.npy'
ekf_outfile = 'ekf_out/zero_noise_output_data.npy'
pf_outfile = 'pf_out/output_data.npy'
```

```python
def load_data(file_path, keys):
    #https://stackoverflow.com/a/58000060/6941952
    return list(map(np.load(file_path).get, keys))
```

```python
def calculate_3sig(covariance):
    return [3 * np.sqrt(covariance[i, i, :]) for i in range(covariance.shape[0])]
```

```python
def plot_one_error(error, fill_area, dt, ax, title):
    ax.plot(error, color='b', label='error')
    ax.fill_between(dt, error - fill_area, error + fill_area, color='r', alpha=0.5, label='$\pm 3\sigma$')
    ax.set_title(title, fontsize=20)
    ax.set_xlabel('time', fontsize=14)
    ax.set_ylabel('error', fontsize=14)
    ax.legend()
```

```python
def plot_errors(errors, sig_3, steps=None):
    
    x, y, theta = errors[:, 0], errors[:, 1], errors[:, 2]
    x_sig3, y_sig3, theta_sig3 = sig_3
    
    if steps is None:
        dt = np.arange(len(x))
    else:
        dt = np.arange(steps)
        
    fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(24, 12))
    
    plot_one_error(x, x_sig3, dt, ax[0], '$\hat{x}-x$')
    plot_one_error(y, y_sig3, dt, ax[1], '$\hat{y} - y$')
    plot_one_error(theta, theta_sig3, dt, ax[2], '$\hat{\Theta} - \Theta$')

    plt.show()
```

```python
time, groundtruth_pose  = load_data(file_path=input_file, keys = ['num_steps', 'real_robot_path'])
```

## EKF Plots

```python
ekf_mean_pose, ekf_cov_pose = load_data(file_path=ekf_outfile, keys=['mean_trajectory', 'covariance_trajectory'])

ekf_errors = ekf_mean_pose - groundtruth_pose
ekf_sigmas_3 = calculate_3sig(covariance=ekf_cov_pose)
plot_errors(errors=ekf_errors, sig_3=ekf_sigmas_3, steps=time)
```

## PF Plots

```python
pf_mean_pose, pf_cov_pose = list(map(np.load(pf_outfile).get, ['mean_trajectory', 'covariance_trajectory']))

pf_errors = pf_mean_pose - groundtruth_pose
pf_sigmas_3 = calculate_3sig(covariance=pf_cov_pose)
plot_errors(errors=pf_errors, sig_3=pf_sigmas_3, steps=time)
```
