```python
import numpy as np
import matplotlib.pyplot as plt

main_keys = ['mean_trajectory', 'covariance_trajectory']
input_file = 'evaluation-input.npy'
normal_ekf_out = 'ekf_out/normal_output_data.npy'
normal_pf_out = 'pf_out/normal_output_data.npy'
zero_noise_ekf_out = 'ekf_out/zero_noise_output_data.npy'
b10_ekf_out = 'ekf_out/b10_output_data.npy'
b2_ekf_out = 'ekf_out/b2_output_data.npy'
part100_pf_out = 'pf_out/100_particles_output_data.npy'
part50_pf_out = 'pf_out/50_particles_output_data.npy'
part10_pf_out = 'pf_out/10_particles_output_data.npy'
over_ekf_out = 'ekf_out/overestimated_output_data.npy'
under_ekf_out = 'ekf_out/underestimated_output_data.npy'
```

```python
def load_data(file_path, keys):
    #https://stackoverflow.com/a/58000060/6941952
    return list(map(np.load(file_path).get, keys))
```

```python
def calculate_3sig(covariance):
    return [3 * np.sqrt(covariance[i, i, :]) for i in range(covariance.shape[0])]
```

```python
def plot_one_error(error, dt, ax, title, fill_area=None, label=None):
    ax.plot(error, color='b' if label is None else None, label='error' if label is None else label)
    if fill_area is not None:
        ax.fill_between(dt, error - fill_area, error + fill_area, color='r', alpha=0.5, label='$\pm 3\sigma$')
    
    ax.set_title(title, fontsize=20)
    ax.set_xlabel('time', fontsize=14)
    ax.set_ylabel('error', fontsize=14)
    
    if label is None:
        ax.legend()
```

```python
def plot_errors(errors, sig_3, steps=None, ax=None, label=None):
    x, y, theta = errors[:, 0], errors[:, 1], errors[:, 2]

    if sig_3 is None:
        sig_3 = [None, None, None]
        
    if steps is None:
        dt = np.arange(len(x))
    else:
        dt = np.arange(steps)
        
    show = False
    if ax is None:
        fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(24, 12))
        show = True
    
    plot_one_error(x, dt, ax[0], title='$\hat{x}-x$', fill_area=sig_3[0], label=label)
    plot_one_error(y, dt, ax[1], title='$\hat{y}-y$', fill_area=sig_3[1], label=label)
    plot_one_error(theta, dt, ax[2], title='$\hat{\Theta}-\Theta$', fill_area=sig_3[2], label=label)
    
    if show:
        plt.show()
```

```python
def main_func(file_path, keys, ax=None, label=None, bounds=True):
    mean_pose, cov_pose = load_data(file_path=file_path, keys=keys)
    errors = mean_pose - groundtruth_pose
    sigmas_3 = None
    if bounds:
        sigmas_3 = calculate_3sig(covariance=cov_pose)
    plot_errors(errors=errors, sig_3=sigmas_3, steps=time, ax=ax, label=label)
```

#### The number of particles decreases

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
main_func(part100_pf_out, keys=main_keys, ax=ax, label='particles=100', bounds=False)
main_func(part50_pf_out, keys=main_keys, ax=ax, label='particles=50', bounds=False)
main_func(part10_pf_out, keys=main_keys, ax=ax, label='particles=10', bounds=False)
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

As the number of particles used in *Particle Filter* decreases, it shows that its localization error decreases too, that's why the performance is detoriorating when only $10$ particles are used. However, a big number of particles isn't always a good option because it requires more computaional resources.

#### The filter noise parameters underestimate or overestimate the true noise parameters

`localization_filter.py` file is changed as follows:
* In case of overestimation:
>`
self._alphas = np.power(alphas, 0.2)
self._Q = beta ** 4
`
* In case of underestimation:
>`
self._alphas = np.power(alphas, 5)
self._Q = np.power(beta, 0.2)
`

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
main_func(normal_ekf_out, keys=main_keys, ax=ax, label='normal case', bounds=False)
main_func(over_ekf_out, keys=main_keys, ax=ax, label='overestimation', bounds=False)
main_func(under_ekf_out, keys=main_keys, ax=ax, label='underestimation', bounds=False)
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

Both cases of overestimation and underestimation results in a great deviation away from the true pose. Thus, the normal case has shown the <u>best</u> estimated localization error in most of the cases.
